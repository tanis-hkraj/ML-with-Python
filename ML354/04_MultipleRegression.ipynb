```python
import pandas as pd
df=pd.read_csv('auto-mpg.data',sep='\s+',na_values='?')
df
```

```python
df.columns=['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin','car-name']
df
```

```python
df.isnull().sum()
```

```python
df.corr()
```

```python
df['horsepower'].fillna(df['horsepower'].mean(),inplace=True)
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['car-name']=le.fit_transform(df['car-name'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(df)
plt.show()
```

```python
import numpy as np
corr=np.corrcoef(df.values.T)
hm=sns.heatmap(corr,annot=True)
plt.show()
```

```python
df.drop(columns=['model_year','cylinders'],inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
X=df.iloc[:,:-1].values
```

```python
Y=df['car-name'].values
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.15,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_std=sc.fit_transform(X_train)
X_test_std=sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train_std,Y_train)
Y_pred=lr.predict(X_test_std)
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(Y_test,Y_pred))
print(mean_absolute_error(Y_test,Y_pred))
```
